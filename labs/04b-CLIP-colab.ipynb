```python
import numpy as np
from transformers import CLIPProcessor, CLIPModel, CLIPTokenizer
import torch
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as T
```

```python
# This cell of code will determine if we have an accelerator for running
# our neural networks.
# mps == Apple Silicon device (MX series of Macbooks)
# cuda == Compute Unified Device Architecture is a toolkit from Nvidia and means we have a GPU
# cpu == Just using the general-purpose CPU for our calculations

if hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
    device = torch.device('mps')
elif torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
print('Using device: {0}'.format(device))
```

```python
# we are loading a transformer neural network (more on this architecture later this term)
# there are three components that we need: the model, the image processor, and the tokenizer
# we'll learn more about tokenization later, for now just know that this 

model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14",
                                  dtype=torch.float16,
                                  device_map="auto")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14",
                                          dtype = torch.float16,
                                          use_fast = False,
                                          clean_up_tokenization_spaces = True,
                                          device_map = "auto")
```

```python
model.eval()
```

```python
# this is a small classification function that returns cosine similarities
# for the supplied labels. 
def classify(image,labels):
    img = Image.open(image)
    inputs = processor(text = labels,
                       images = img, 
                       return_tensors = "pt",
                       padding = True) 
    outputs = model(**inputs.to(device))
    logits = outputs.logits_per_image
    
    # we're using softmax to render our similarities in the 0-1 scale and ensure
    # that the returned values, the logits, sum to 1. This creates dependencies on our
    # input list of possible labels (text inputs). We can take raw logits but they 
    # won't be scaled and summed to 1.
    probs = logits.softmax(dim = 1)  
    
    # fix for float16 data
    return [(labels[c],np.round(probs[0][c].detach().to('cpu').numpy(),3)) 
            for c in torch.argsort(probs,descending=True)[0]][:10]
```

```python
def classify(image, labels):
    img = Image.open(image)
    inputs = processor(text=labels,
                       images=img, 
                       return_tensors="pt",
                       padding=True) 
    outputs = model(**inputs.to(device))
    logits = outputs.logits_per_image  # These are already cosine similarities (0-100 scale)
    
    # Use raw logits instead of softmax
    return [(labels[c], np.round(logits[0][c].detach().to('cpu').numpy(), 3)) 
            for c in torch.argsort(logits, descending=True)[0]][:10]
```
