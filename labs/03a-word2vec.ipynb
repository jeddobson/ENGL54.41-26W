```python
import gensim
from gensim.models import Word2Vec
from gensim.models import KeyedVectors

from gensim import matutils
import numpy as np
from numpy import dot
from sklearn.metrics.pairwise import cosine_similarity

from sklearn.decomposition import PCA
from sklearn.manifold import MDS

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# we need to install this package in order to download files from a Google Drive
!pip install gdown > /dev/null 2>&1
import gdown
```

```python
# now download the Google News 300 model (1.65GB compressed)
url = "https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit?usp=sharing"
binary_model = "/home/jovyan/GoogleNews-vectors-negative300.bin.gz"
gdown.download(fuzzy=True, output=binary_model, url=url)
```

```python
# Load the model --- this will take time!
google_model = KeyedVectors.load_word2vec_format(binary_model,
                                                 binary = True)
```

```python
# "Interview" the model
vocab_size, dim = google_model.vectors.shape
print("vocab:", vocab_size)
print("depth:", dim)
```

```python
# note that this is a LARGE model and it has bigrams and a mix of
# capitalized and uncapitalized words. We'll also be using a smaller model without
# these particular features in the vocabulary.
```

```python
google_model["tuesday"]
```

```python
google_model.most_similar(["apple","microsoft"])
```

```python
# Similar Terms
google_model.most_similar("Dartmouth",topn=25)
```

```python
# What is the distance (cosine similarity, so angle similarity) between dartmouth and hanover?
google_model.distance('Dartmouth','Hanover')
```

```python
# Similar Terms: We can group terms together in a list to search for neighbors 
# of this ("concept?") group.
google_model.most_similar(["Dartmouth","Harvard","Yale"],topn=25)
```

```python
# This example was introduced with the model
google_model.most_similar(positive=['woman', 'king'], negative=['man'])
```

```python
# here's another version of this query
google_model.most_similar(positive=['Oslo', 'Portugal'], negative=['Lisbon'])
```

```python
# and our famous query!
google_model.most_similar(positive=['woman', 'computer_programmer'], negative=['man'])
```
